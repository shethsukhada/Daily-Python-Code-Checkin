```python
import pandas as pd
#https://www.programiz.com/python-programming/datetime
```

```python
date_df = pd.read_excel('dates.xlsx',parse_dates=['Date1','Date2','Date3'])
```

```python
date_df.head()
```

```python
date_df.tail()
```

```python
def extract_date_parts(pd,column):    
    pd['month_'+column] = pd[column].apply(lambda x:x.month)
    pd['day_'+column] = pd[column].apply(lambda x:x.day)
```

```python
extract_date_parts(date_df,'Date1')
extract_date_parts(date_df,'Date2')
extract_date_parts(date_df,'Date3')
```
